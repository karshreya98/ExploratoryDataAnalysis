```python
#import csv files
import pandas as pd
test=pd.read_csv('Kaggle_Competition/santander-customer-transaction-prediction/test.csv')
train=pd.read_csv('Kaggle_Competition/santander-customer-transaction-prediction/train.csv')
```

```python
#correlation removal
train.corr()
```

```python
cols=list(train.columns.values)
cols=cols[2:]
cols
```

```python
from sklearn.metrics import roc_auc_score
```

```python
import lightgbm as lgb
target=train['target']
```

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=15, shuffle=False)
oof = np.zeros(len(train))
predictions = np.zeros(len(test))
param = {
    'bagging_freq': 5,          
    'bagging_fraction': 0.331,   'boost_from_average':'false',   
    'boost': 'gbdt',             'feature_fraction': 0.0405,     'learning_rate': 0.0083,
    'max_depth': -1,             'metric':'auc',                'min_data_in_leaf': 80,     'min_sum_hessian_in_leaf': 10.0,
    'num_leaves': 13,            'num_threads': 8,              'tree_learner': 'serial',   'objective': 'binary',      'verbosity': 1
}
for fold_, (trn_idx, val_idx) in enumerate(folds.split(train.values, target.values)):
    print("Fold {}".format(fold_))
    trn_data = lgb.Dataset(train.iloc[trn_idx][cols], label=target.iloc[trn_idx])
    val_data = lgb.Dataset(train.iloc[val_idx][cols], label=target.iloc[val_idx])
    clf = lgb.train(param, trn_data, 1000000, valid_sets = [trn_data, val_data], verbose_eval=5000, early_stopping_rounds = 4000)
    oof[val_idx] = clf.predict(train.iloc[val_idx][cols], num_iteration=clf.best_iteration)
    predictions += clf.predict(test[cols], num_iteration=clf.best_iteration) / folds.n_splits
print("CV score: {:<8.5f}".format(roc_auc_score(target, oof)))
```
